```python
#！设置提取深度
#引入头文件
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import os
#定义深度,按照0.5km为间距进行取样
depth_spacing = 0.5
#最大深度150km
depth_max = 150
#设定提取的每个深度
depth = np.arange(0,depth_max+depth_spacing,depth_spacing)
print("depth:",depth)
```

```python
#!设置提取深度的文件名
#？设定depth_visual的目的是什么？
depth_visual = np.array([0,5,10,15,20,30,40,60,80,100,120,depth_max])
#设置提取文件的横向值和竖向值
#文件横向值从28.00开始到49.00结束，竖向值从200（底）到260结束
#count用来计数文件的个数
count = 0
#？设定lay1到lay13的目的是什么？
lay1 = [];lay2 = [];lay3 = [];lay4 = [];lay5 = [];lay6 = [];lay7 = [];lay8 = [];lay9 = [];
lay10 = [];lay11 = [];lay12 = [];lay13 = [];
for lat in np.arange(28,50,0.1):
    for lon in np.arange(200,260+0.1,0.1):
        #定义文件的名字
        key = '%.2f'%lat + "_" + "%.2f"%lon
        key_name = key + ".txt"
        #定义文件的路径
        #?路径还没有设置好
        Vs_path = './InputVsData/'
        Vs_file = Vs_path + key_name
        if os.path.exists(Vs_file):
            #引入路径文件
            temp = np.loadtxt(Vs_file)
            if len(temp) > 1:
                count = count + 1
                print(count,key_name)
                #读取文件的第一列数据&第二列数据
                depth_temp = temp[:,0]
                vs_temp = temp[:,1]
                #用这些数据去和CNN的训练结果进行比较
                #线性插值，返回线性插值的函数
                f1 = interp1d(depth_temp,vs_temp,kind = 'slinear')
                #把预设的固定深度带入到线性插值的函数中
                vs = f1(depth)
                #把深度和计算出的Vs组成矩阵
                vs_out = np.array([depth,vs])
                #把预设深度0-150km，按照每0.5km进行进行取值，把结果存入到Vs_Inversional_Truth文件夹中
                #?vs_out.T是什么意思?
                np.savetxt('./Vs_Inversional_Truth/'+key_name,vs_out.T,fmt="%10.5f")
                #按照设定的新的深度值，0,5,10,15,20,30,40,60,80,100,120,150km对f1函数重新取一遍值
                vs = f1(depth_visual)
                #把文件中的横向值、竖向值、和vs中的第i个参数进行保存
                #不同深度的数据保存在不同的层矩阵（layi）中
                lay1.append([lon,lat,vs[0]])
                lay2.append([lon,lat,vs[1]])
                lay3.append([lon,lat,vs[2]])
                lay4.append([lon,lat,vs[3]])
                lay5.append([lon,lat,vs[4]])
                lay6.append([lon,lat,vs[5]])
                lay7.append([lon,lat,vs[6]])
                lay8.append([lon,lat,vs[7]])
                lay9.append([lon,lat,vs[8]])
                lay10.append([lon,lat,vs[9]])
                lay11.append([lon,lat,vs[10]])
                lay12.append([lon,lat,vs[11]])
```
